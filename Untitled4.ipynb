```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider
import seaborn as sns

df = pd.read_csv('weather.csv')
df['time'] = pd.to_datetime(df['time'])
df['year'] = df['time'].dt.year
df['month'] = df['time'].dt.month
df['Ftemp'] = (df['Ktemp'] - 273.15) * (9/5) + 32

monthly_avg = df.groupby(['year', 'month'])['Ftemp'].mean().reset_index()
years = sorted(monthly_avg['year'].unique())
initial_year = years[0]

plt.style.use('seaborn-v0_8-darkgrid')
fig, ax = plt.subplots(figsize=(12, 7), facecolor='#f0f0f0')
plt.subplots_adjust(bottom=0.25)

year_data = monthly_avg[monthly_avg['year'] == initial_year]
line, = ax.plot(year_data['month'], year_data['Ftemp'],
                marker='o',
                linestyle='-',
                linewidth=3,
                markersize=8,
                color='#1E90FF',
                markeredgecolor='navy',
                markerfacecolor='lightskyblue')

ax.set_facecolor('#F5F5F5')
ax.set_xlabel('Month', fontsize=14, fontweight='bold')
ax.set_ylabel('Temperature (°F)', fontsize=14, fontweight='bold')
ax.set_title(f'Average Monthly Temperature (°F) - {initial_year}',
             fontsize=16,
             fontweight='bold',
             color='#333333')
ax.set_xticks(range(1, 13))
ax.set_xticklabels(['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
                    'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
                   fontweight='bold')
ax.grid(True, linestyle='--', alpha=0.7)

ymin = monthly_avg['Ftemp'].min() - 5
ymax = monthly_avg['Ftemp'].max() + 5
ax.set_ylim(ymin, ymax)

ax_slider = plt.axes([0.2, 0.1, 0.65, 0.03],
                     facecolor='lightsteelblue')
year_slider = Slider(
    ax=ax_slider,
    label='Year',
    valmin=0,
    valmax=len(years) - 1,
    valinit=0,
    valstep=1,
    color='#4169E1',
    track_color='lightgray'
)

def update(val):
    year_idx = int(year_slider.val)
    selected_year = years[year_idx]
    year_data = monthly_avg[monthly_avg['year'] == selected_year]
    line.set_xdata(year_data['month'])
    line.set_ydata(year_data['Ftemp'])
    ax.set_title(f'Average Monthly Temperature (°F) - {selected_year}',
                 fontsize=16,
                 fontweight='bold',
                 color='#333333')
    fig.canvas.draw_idle()

year_slider.on_changed(update)

ax_slider.set_xticks(range(len(years)))
ax_slider.set_xticklabels([str(year) for year in years],
                           rotation=45,
                           fontsize=8,
                           fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

yearly_avg = df.groupby('year')['Ftemp'].mean().reset_index()

warm_years = yearly_avg[yearly_avg['Ftemp'] > 55]
first_warm_year = warm_years.iloc[0]['year']
first_warm_temp = warm_years.iloc[0]['Ftemp']
print(f"\nThe first year with an average temperature above 55°F is {first_warm_year} with {first_warm_temp:.2f}°F")


plt.figure(figsize=(12, 6))
plt.plot(yearly_avg['year'], yearly_avg['Ftemp'], 'o-', linewidth=2)
plt.axhline(y=55, color='r', linestyle='--', label='55°F Threshold')

plt.scatter(first_warm_year, first_warm_temp, color='red', s=100, zorder=5)
plt.annotate(f"{first_warm_year}: {first_warm_temp:.2f}°F",
                 xy=(first_warm_year, first_warm_temp),
                 xytext=(first_warm_year + 0.5, first_warm_temp + 1),
                 fontsize=12)

plt.title('Yearly Average Temperature (°F)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Temperature (°F)', fontsize=12)
plt.grid(True)
plt.legend()
plt.tight_layout()

plt.savefig('yearly_temperature_analysis.png', dpi=300)

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import matplotlib.colors as mcolors
from matplotlib.colors import LinearSegmentedColormap
import seaborn as sns
import calendar
from datetime import datetime
from matplotlib.cm import ScalarMappable
from scipy import stats

temp_df = pd.read_csv('weather.csv')
temp_df['time'] = pd.to_datetime(temp_df['time'])
temp_df['year'] = temp_df['time'].dt.year
temp_df['month'] = temp_df['time'].dt.month
temp_df['month_name'] = temp_df['month'].apply(lambda x: calendar.month_abbr[x])
temp_df['Ftemp'] = (temp_df['Ktemp'] - 273.15) * (9/5) + 32

rainfall_df = pd.read_csv('rainfall.csv')
months = ['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']
rainfall_long = pd.melt(
    rainfall_df,
    id_vars=['YEAR'],
    value_vars=months,
    var_name='MONTH',
    value_name='RAINFALL'
)

month_map = {month: i+1 for i, month in enumerate(months)}
rainfall_long['month'] = rainfall_long['MONTH'].map(month_map)
rainfall_long['month_name'] = rainfall_long['month'].apply(lambda x: calendar.month_abbr[x])
rainfall_long.rename(columns={'YEAR': 'year'}, inplace=True)

monthly_temp = temp_df.groupby(['year', 'month']).agg({
    'Ftemp': 'mean'
}).reset_index()

merged_data = pd.merge(
    monthly_temp,
    rainfall_long,
    on=['year', 'month'],
    how='inner'
)

correlation, p_value = stats.pearsonr(merged_data['Ftemp'], merged_data['RAINFALL'])

def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

merged_data['season'] = merged_data['month'].apply(get_season)

season_corr = {}
for season in merged_data['season'].unique():
    season_data = merged_data[merged_data['season'] == season]
    corr, p = stats.pearsonr(season_data['Ftemp'], season_data['RAINFALL'])
    season_corr[season] = (corr, p)

merged_data['date'] = pd.to_datetime(merged_data[['year', 'month']].assign(day=15))
merged_data = merged_data.sort_values('date')

fig1, ax1 = plt.subplots(figsize=(15, 8))

color = 'tab:red'
ax1.set_xlabel('Year', fontsize=12)
ax1.set_ylabel('Temperature (°F)', color=color, fontsize=12)
ax1.plot(merged_data['date'], merged_data['Ftemp'], color=color, alpha=0.7)
ax1.tick_params(axis='y', labelcolor=color)
ax1.grid(True, alpha=0.3)

ax2 = ax1.twinx()
color = 'tab:blue'
ax2.set_ylabel('Monthly Rainfall (inches)', color=color, fontsize=12)
ax2.bar(merged_data['date'], merged_data['RAINFALL'], color=color, alpha=0.4, width=25)
ax2.tick_params(axis='y', labelcolor=color)

ax1.xaxis.set_major_locator(mdates.YearLocator(5))
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

plt.title('Monthly Temperature and Rainfall Over Time', fontsize=16)
fig1.tight_layout()
plt.savefig('temp_rainfall_timeseries.png', dpi=300, bbox_inches='tight')
plt.close(fig1)

fig2, ax3 = plt.subplots(figsize=(12, 10))

years = merged_data['year'].unique()
norm = mcolors.Normalize(vmin=min(years), vmax=max(years))
cmap = plt.cm.viridis

scatter = ax3.scatter(
    merged_data['Ftemp'],
    merged_data['RAINFALL'],
    c=merged_data['year'],
    cmap=cmap,
    norm=norm,
    s=50,
    alpha=0.7,
    edgecolor='k',
    linewidth=0.5
)

cbar = fig2.colorbar(scatter, ax=ax3)
cbar.set_label('Year', fontsize=12)

m, b = np.polyfit(merged_data['Ftemp'], merged_data['RAINFALL'], 1)
ax3.plot(
    merged_data['Ftemp'],
    m * merged_data['Ftemp'] + b,
    'r--',
    linewidth=2,
    label=f'Correlation: {correlation:.2f}'
)

ax3.set_xlabel('Temperature (°F)', fontsize=14)
ax3.set_ylabel('Monthly Rainfall (inches)', fontsize=14)
ax3.set_title('Temperature vs. Rainfall\nColor Gradient Shows Year', fontsize=16)
ax3.grid(True, alpha=0.3)
ax3.legend()

plt.tight_layout()
plt.savefig('temp_rainfall_scatter.png', dpi=300, bbox_inches='tight')
plt.close(fig2)

fig3, axs = plt.subplots(2, 2, figsize=(15, 10))
fig3.suptitle('Seasonal Temperature-Rainfall Relationship', fontsize=16)

seasons = ['Winter', 'Spring', 'Summer', 'Fall']
colors = ['#1e90ff', '#2ecc71', '#e74c3c', '#f39c12']

for i, season in enumerate(seasons):
    row, col = i // 2, i % 2
    ax = axs[row, col]

    season_data = merged_data[merged_data['season'] == season]

    scatter = ax.scatter(
        season_data['Ftemp'],
        season_data['RAINFALL'],
        c=season_data['year'],
        cmap='viridis',
        s=60,
        alpha=0.7,
        edgecolor='k',
        linewidth=0.5
    )

    if len(season_data) > 1:
        m, b = np.polyfit(season_data['Ftemp'], season_data['RAINFALL'], 1)
        x_range = np.linspace(season_data['Ftemp'].min(), season_data['Ftemp'].max(), 100)
        ax.plot(x_range, m * x_range + b, '--', color=colors[i], linewidth=2)

        corr_val, p_val = season_corr[season]
        ax.text(
            0.05, 0.95,
            f'r = {corr_val:.2f}\np = {p_val:.4f}',
            transform=ax.transAxes,
            fontsize=10,
            verticalalignment='top',
            bbox=dict(boxstyle='round', facecolor='white', alpha=0.7)
        )

    ax.set_title(f'{season}', fontsize=14, color=colors[i])
    ax.set_xlabel('Temperature (°F)', fontsize=12)
    ax.set_ylabel('Monthly Rainfall (inches)', fontsize=12)
    ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.subplots_adjust(top=0.9)
plt.savefig('seasonal_temp_rainfall.png', dpi=300, bbox_inches='tight')
plt.close(fig3)

fig4, ax4 = plt.subplots(figsize=(12, 8))

monthly_climate = merged_data.groupby('month').agg({
    'Ftemp': 'mean',
    'RAINFALL': 'mean',
    'month_name': 'first'
}).reset_index()

monthly_climate = monthly_climate.sort_values('month')

for i in range(len(monthly_climate)):
    month = monthly_climate.iloc[i]['month_name']
    x = monthly_climate.iloc[i]['Ftemp']
    y = monthly_climate.iloc[i]['RAINFALL']

    ax4.scatter(x, y, s=120, color=plt.cm.hsv(i/12), edgecolor='black', linewidth=1, zorder=3)

    ax4.annotate(
        month,
        xy=(x, y),
        xytext=(0, 0),
        textcoords='offset points',
        fontsize=12,
        weight='bold',
        ha='center',
        va='center',
        color='white'
    )

    if i < len(monthly_climate) - 1:
        next_x = monthly_climate.iloc[i+1]['Ftemp']
        next_y = monthly_climate.iloc[i+1]['RAINFALL']
        ax4.arrow(
            x, y,
            next_x - x, next_y - y,
            head_width=0.1,
            head_length=0.1,
            fc='gray',
            ec='gray',
            alpha=0.6,
            length_includes_head=True,
            zorder=1
        )

ax4.set_xlabel('Average Temperature (°F)', fontsize=14)
ax4.set_ylabel('Average Rainfall (inches)', fontsize=14)
ax4.set_title('Monthly Temperature-Rainfall Cycle', fontsize=16)
ax4.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('monthly_temp_rainfall_cycle.png', dpi=300, bbox_inches='tight')
plt.close(fig4)

temp_pivot = merged_data.pivot(index='year', columns='month', values='Ftemp')
rainfall_pivot = merged_data.pivot(index='year', columns='month', values='RAINFALL')

fig5, (ax5, ax6) = plt.subplots(2, 1, figsize=(15, 12))

sns.heatmap(
    temp_pivot,
    cmap='RdBu_r',
    ax=ax5,
    cbar_kws={'label': 'Temperature (°F)'},
    xticklabels=[calendar.month_abbr[i] for i in range(1, 13)]
)
ax5.set_title('Monthly Temperatures by Year', fontsize=14)
ax5.set_ylabel('Year', fontsize=12)

sns.heatmap(
    rainfall_pivot,
    cmap='YlGnBu',
    ax=ax6,
    cbar_kws={'label': 'Rainfall (inches)'},
    xticklabels=[calendar.month_abbr[i] for i in range(1, 13)]
)
ax6.set_title('Monthly Rainfall by Year', fontsize=14)
ax6.set_xlabel('Month', fontsize=12)
ax6.set_ylabel('Year', fontsize=12)

plt.tight_layout()
plt.savefig('temp_rainfall_heatmap.png', dpi=300, bbox_inches='tight')
plt.close(fig5)

monthly_climate = monthly_climate.sort_values('month')
rainiest_month = monthly_climate.loc[monthly_climate['RAINFALL'].idxmax()]
driest_month = monthly_climate.loc[monthly_climate['RAINFALL'].idxmin()]
warmest_month = monthly_climate.loc[monthly_climate['Ftemp'].idxmax()]
coolest_month = monthly_climate.loc[monthly_climate['Ftemp'].idxmin()]

print(f"Overall correlation: {correlation:.4f} (p-value: {p_value:.4f})")
for season in season_corr:
    print(f"{season} correlation: {season_corr[season][0]:.4f} (p-value: {season_corr[season][1]:.4f})")
print(f"Rainiest: {rainiest_month['month_name']} ({rainiest_month['RAINFALL']:.2f}in)")
print(f"Driest: {driest_month['month_name']} ({driest_month['RAINFALL']:.2f}in)")
print(f"Warmest: {warmest_month['month_name']} ({warmest_month['Ftemp']:.1f}°F)")
print(f"Coolest: {coolest_month['month_name']} ({coolest_month['Ftemp']:.1f}°F)")
```
